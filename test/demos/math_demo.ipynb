## Tensor Contractions

```julia
f, ctx = @pct begin
    @space ST begin
        type=(N,N,N) -> R
        symmetries=(((2,1,3), :id), ((1,3,2), :id))
    end

    (T::ST) -> pullback((x::RV) -> sum((i, j, k), T(i, j, k) * x(i) * x(j) * x(k)))
end; println(latex(f))
```

```julia
get_body(get_body(f)) |> decompose 
```

```julia
df = redux(vdiff(f); settings=symmetry_settings)
```

```julia
println(latex(df))
```

```julia
using MacroTools
println(MacroTools.striplines(codegen(df)))
```

## Hartree-Fock Contraction (Complex Numbers)

```julia
_, ctx = @pct begin    
    @space T begin
        type = (I, I, I, I) -> C
        symmetries = (((2, 1, 4, 3), :conj), ((3, 4, 1, 2), :id))
    end
end;
```

```julia
f, _ = @pct _ ctx (J::T) -> pullback((C::CM) -> sum((i, j, p, q, r, s),
    C(p, i)' * C(q, i) * C(r, j)' * C(s, j) * J(p, q, r, s))); f
```

```julia
df = redux(vdiff(f); settings=symmetry_settings)
```

## Deriving the Conjugate Gradient

```julia
f, _ = @pct (A::Sym, r::RV, p::RV, b::RV, x::RV) -> begin
    E = (x::RV) -> sum((i, j), 0.5 * x(i) * A(i, j) * x(j)) - sum(i, x(i) * b(i))
    pullback((Î±::R, Î²::R) -> E((i::N) -> x(i) + Î± * (r(i) + Î² * p(i))))
end; blaserize(f)
```
